```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV   
csv_file_path_1 = "expr1_ep_length.csv"
csv_file_path_2 = "expr2_ep_length.csv"

# CSV  
data_1 = pd.read_csv(csv_file_path_1)
data_2 = pd.read_csv(csv_file_path_2)

#    
window_size = 40

# 'Step' 'Value'     
x1 = data_1['Step']
y1 = data_1['Value'].rolling(window=window_size).mean()

x2 = data_2['Step']
y2 = data_2['Value'].rolling(window=window_size).mean()

#     
plt.figure(figsize=(12, 7))

plt.plot(x1, y1, color='red', label='Original Loss')
plt.plot(x2, y2, color='blue', label='Time-Weighted Loss')

#     (  )
plt.xlabel('Timesteps', fontsize=23)
plt.ylabel('Episode Lengths', fontsize=23)
plt.title('Original Loss <EP_Length> vs Time-Weighted Loss <EP_Length>', fontsize=18, pad=20)
plt.legend(fontsize=20)

#   
plt.minorticks_on()

#   (  +  )
plt.grid(True, which='major', linestyle='--', alpha=0.7)
plt.grid(True, which='minor', linestyle=':', alpha=0.4)

#     
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

#    
ax = plt.gca()
ax.ticklabel_format(style='sci', axis='x')  #   
ax.xaxis.offsetText.set_fontsize(18)        # 1e8    

#  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#  
def smooth(data, smoothing_factor=0.99):
    smoothed_data = np.zeros_like(data)
    smoothed_data[0] = data[0]
    for i in range(1, len(data)):
        smoothed_data[i] = smoothing_factor * smoothed_data[i-1] + (1 - smoothing_factor) * data[i]
    return smoothed_data

# CSV   
files = [
    "LoRa_type1.csv",
    "LoRa_type2.csv",
    "LoRa_type3.csv"
]

#  CSV       
plt.figure(figsize=(10, 6))

for file in files:
    df = pd.read_csv(file)
    
    # TensorBoard  'step' 'value'   
    if 'Step' in df.columns and 'Value' in df.columns:
        #  
        smoothed_values = smooth(df['Value'].values, smoothing_factor=0.99)
        
        #  
        plt.plot(df['Step'], smoothed_values, label=file.split('.')[0])

#    
plt.title("LoRa Efficient SF allocation")
plt.xlabel("Timestep")
plt.ylabel("Packet Collision Rate")
plt.legend()

#  
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
x_labels = ['SF7', 'SF8', 'SF9', 'SF10', 'SF11', 'SF12']
type1 = [0.1814, 0.3615, 0.3334, 0.0892, 0.0197, 0.0148]
type2 = [0.2296, 0.1848, 0.314, 0.1869, 0.0471, 0.0376]
type3 = [0.6562, 0.2352, 0.0568, 0.0181, 0.0157, 0.018]

#  
sum_type1 = sum(type1)
sum_type2 = sum(type2)
sum_type3 = sum(type3)

type1 = [x / sum_type1 for x in type1]
type2 = [x / sum_type2 for x in type2]
type3 = [x / sum_type3 for x in type3]

#  
x = np.arange(len(x_labels))

#  
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.25
ax.bar(x - bar_width, type1, bar_width, label='Type1', color='blue')
ax.bar(x, type2, bar_width, label='Type2', color='orange')
ax.bar(x + bar_width, type3, bar_width, label='Type3', color='green')

#  
ax.set_xlabel('SF value')
ax.set_ylabel('Allocation Ratio')
ax.set_title('LoRa Efficient SF allocation')
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.legend()

#  
plt.tight_layout()
plt.show()
```
